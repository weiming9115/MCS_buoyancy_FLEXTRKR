```python
def dBL_component_map(BL_TOT_climat, BL_CAPE_climat, BL_SUBSAT_climat,
                      BL_MCSint):
    
    dBL_comps_list = []
    
    for month_sel in np.arange(1,13):
    
        BL_TOT_ref = (340*3/9.81)*BL_TOT_climat.sel(month=month_sel).sel(lat=slice(-30,30))
        BL_CAPE_ref = (340*3/9.81)*BL_CAPE_climat.sel(month=month_sel).sel(lat=slice(-30,30))
        BL_SUBSAT_ref = (340*3/9.81)*BL_SUBSAT_climat.sel(month=month_sel).sel(lat=slice(-30,30)) 

        BL_TOT_diff = (BL_MCSint.buoy_tot_int/BL_MCSint.samples).sel(month=month_sel) - BL_TOT_ref
        BL_TOT_diff = BL_TOT_diff.rename('dBL_TOT')
        BL_CAPE_diff = (BL_MCSint.buoy_cape_int/BL_MCSint.samples).sel(month=month_sel) - BL_CAPE_ref
        BL_CAPE_diff = BL_CAPE_diff.rename('dBL_CAPE')
        BL_SUBSAT_diff = (BL_MCSint.buoy_subsat_int/BL_MCSint.samples).sel(month=month_sel) - BL_SUBSAT_ref
        BL_SUBSAT_diff = BL_SUBSAT_diff.rename('dBL_SUBSAT')
                
        # rename ref vars
        BL_TOT_ref = BL_TOT_ref.to_dataset().rename({'Buoy_TOT':'BL_TOT_ref'})
        BL_CAPE_ref = BL_CAPE_ref.to_dataset().rename({'Buoy_CAPE':'BL_CAPE_ref'})
        BL_SUBSAT_ref = BL_SUBSAT_ref.to_dataset().rename({'Buoy_SUBSAT':'BL_SUBSAT_ref'})        
            
        ds = xr.merge([BL_TOT_ref, BL_CAPE_ref, BL_SUBSAT_ref, BL_TOT_diff, BL_CAPE_diff, BL_SUBSAT_diff])
        dBL_comps_list.append(ds)
        
    dBL_stats = xr.concat(dBL_comps_list, pd.Index(np.arange(1,13),name='month'))
    dBL_stats = xr.merge([dBL_stats, BL_MCSint.samples])
    
    return dBL_stats
```

```python
dBL_stats_monthly = dBL_component_map(BL_TOT_climat, BL_CAPE_climat, BL_SUBSAT_climat, BL_TOT_MCSint)
```

#### Regional diagnostics

```python
def regional_BL_diag(BL_MCS, BL_Climat, lat_range, lon_range):
    """
    return regional BL-MCS and BL-CLimat components
    """
    BL_MCS_region = BL_MCS.sel(lat=slice(lat_range[0], lat_range[-1]),
                               lon=slice(lon_range[0], lon_range[-1]))
    BL_Climat_region = BL_Climat.sel(lat=slice(lat_range[0], lat_range[-1]),
                               lon=slice(lon_range[0], lon_range[-1]))
    
    # calculate the difference at the grid scale
    BL_TOT_diff = (BL_MCS_region.buoy_tot_int)/(BL_MCS_region.samples)  - BL_Climat_region.Buoy_TOT # (K)
    BL_CAPE_diff = (BL_MCS_region.buoy_cape_int)/(BL_MCS_region.samples) - BL_Climat_region.Buoy_CAPE # (K)
    BL_SUBSAT_diff = (BL_MCS_region.buoy_subsat_int)/(BL_MCS_region.samples) - BL_Climat_region.Buoy_SUBSAT # (K)
    
    return BL_TOT_diff, BL_CAPE_diff, BL_SUBSAT_diff, BL_Climat_region, BL_MCS_region.samples
```
