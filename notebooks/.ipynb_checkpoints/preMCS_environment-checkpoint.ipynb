```python
def es_calc_bolton(temp):
    # in hPa

    tmelt  = 273.15
    tempc = temp - tmelt 
    es = 6.112*np.exp(17.67*tempc/(243.5+tempc))
    
    return es
```

```python
def es_calc(temp):

    tmelt  = 273.15
    
    c0=0.6105851e+03
    c1=0.4440316e+02
    c2=0.1430341e+01
    c3=0.2641412e-01
    c4=0.2995057e-03
    c5=0.2031998e-05
    c6=0.6936113e-08
    c7=0.2564861e-11
    c8=-.3704404e-13

    tempc = temp - tmelt 
    tempcorig = tempc
        
    #if tempc < -80:
    es_ltn80c = es_calc_bolton(temp)
    es_ltn80c = es_ltn80c.where(tempc < -80, 0)
    #else:
    es = c0+tempc*(c1+tempc*(c2+tempc*(c3+tempc*(c4+tempc*(c5+tempc*(c6+tempc*(c7+tempc*c8)))))))
    es_gtn80c = es/100
    es_gtn80c = es_gtn80c.where(tempc >= -80, 0)
    
    # complete es
    es = es_ltn80c + es_gtn80c
    
    return es
```

```python
def qs_calc(temp):

    tmelt  = 273.15
    RV=461.5
    RD=287.04

    EPS=RD/RV

    press = temp.level * 100. # in Pa
    tempc = temp - tmelt 

    es=es_calc(temp) 
    es=es * 100. #hPa
    qs = (EPS * es) / (press + ((EPS-1.)*es))
    
    return qs
```

```python
def theta_e_calc(temp, q):
    
    # following the definitions in Bolton (1980): the calculation of equivalent potential temperature
    
    pref = 100000.
    tmelt  = 273.15
    CPD=1005.7
    CPV=1870.0
    CPVMCL=2320.0
    RV=461.5
    RD=287.04
    EPS=RD/RV
    ALV0=2.501E6

    press = temp.level * 100. # in Pa
    tempc = temp - tmelt # in C

    r = q / (1. - q) 

    # get ev in hPa 
    ev_hPa = temp.level * r / (EPS + r) # hpa

    #get TL
    TL = (2840. / ((3.5*np.log(temp)) - (np.log(ev_hPa)) - 4.805)) + 55.

    #calc chi_e:
    chi_e = 0.2854 * (1. - (0.28 * r))

    theta_e = temp * np.power((pref / press),chi_e) * np.exp(((3.376/TL) - 0.00254) * r * 1000 * (1. + (0.81 * r)))
    
    return theta_e
```

```python
dir_mcs_env = Path('/neelin2020/mcs_flextrkr/era5_envs')
dir_non2mcs = Path('/neelin2020/mcs_flextrkr/mcs_stats')

# load non2mcs tracks
year = 2014
data_non2mcs_complete = xr.open_dataset(dir_non2mcs / 'mcs_tracks_non2mcs_{}.IndoPacfic.nc'.format(year))
data_mcsenvs = xr.open_dataset(dir_mcs_env / 'mcs_era5_mean_envs_{}0101.0000_{}0101.0000.nc'.format(year,year+1))

# extract filtered tracks
data_mcsenvs = data_mcsenvs.sel(tracks=data_non2mcs_complete.tracks)

# load processed rainrate data
data_rainrate = xr.open_dataset(dir_mcs_env / 'mcs_era5_non2mcs_rainrate_2014.IndoPacfic.nc')
```

```python
### classify tracks by their max precip into quantiles
rainmax = data_non2mcs_complete.total_rain.max(dim='times')

idx_quantile = []
qt = [0,0.25,0.5,0.75,1]
for i in range(len(qt)-1):
     
    tmp = rainmax.where((rainmax >= rainmax.quantile(qt[i])) & (rainmax < rainmax.quantile(qt[i+1]))).dropna(dim='tracks')
    idx_quantile.append(tmp.tracks) # save tracks corresponding to the specific quantiles

    
data_non2mcs_q1 = data_non2mcs_complete.sel(tracks=idx_quantile[0])
data_non2mcs_q2 = data_non2mcs_complete.sel(tracks=idx_quantile[1])
data_non2mcs_q3 = data_non2mcs_complete.sel(tracks=idx_quantile[2])
data_non2mcs_q4 = data_non2mcs_complete.sel(tracks=idx_quantile[3])

# environmental variables 
data_non2mcs_envs_q1 = data_rainrate.sel(tracks=idx_quantile[0])
data_non2mcs_envs_q2 = data_rainrate.sel(tracks=idx_quantile[1])
data_non2mcs_envs_q3 = data_rainrate.sel(tracks=idx_quantile[2])
data_non2mcs_envs_q4 = data_rainrate.sel(tracks=idx_quantile[3])

data_non2mcs_pr_q1 = data_rainrate.sel(tracks=idx_quantile[0])
data_non2mcs_pr_q2 = data_rainrate.sel(tracks=idx_quantile[1])
data_non2mcs_pr_q3 = data_rainrate.sel(tracks=idx_quantile[2])
data_non2mcs_pr_q4 = data_rainrate.sel(tracks=idx_quantile[3])
```

```python
# changes in omega profiles - see if a shallow upwelling starts before the initiation

fig,(ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(6,6))

for ax,i in zip([ax1,ax2,ax3,ax4],range(4)):
    tmp = data_mcsenvs['Q'].sel(tracks=idx_quantile[i].values, rel_times=slice(-23,0)).mean('tracks')
    cf = ax.contourf(tmp.rel_times, tmp.level, 1000*tmp.T-1000*tmp.T.sel(rel_times=-23),
               levels=np.linspace(0,1,11), cmap='jet_r', extend='both')
    ax.contour(tmp.rel_times, tmp.level, 1000*tmp.T-1000*tmp.T.sel(rel_times=-23), 
               levels=np.linspace(0,1,11), colors=['k'], linewidths=0.5)
    plt.colorbar(cf, ax=ax)
    if i == 0:
        ax.set_title('vertical profile of $\Delta$qv (before t=0 of MCSs)',fontsize=10)
    ax.set_xlabel('Time prior to init. (hr)')   
    ax.set_ylabel('Pressure (hPa)')
    ax.set_yticks([1000,850,500,200])
    ax.set_ylim([1000,200])
```

```python
# plot evolution of equivalent potential temperature

fig,(ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(6,6))

for ax,i in zip([ax1,ax2,ax3,ax4],range(4)):
    
    q = data_mcsenvs['Q'].sel(tracks=idx_quantile[i].values, rel_times=slice(-23,0)).mean('tracks')
    T = data_mcsenvs['T'].sel(tracks=idx_quantile[i].values, rel_times=slice(-23,0)).mean('tracks')
    tmp = theta_e_calc(T, q) # thetae profile
    cf = ax.contourf(tmp.rel_times, tmp.level, tmp.T-tmp.T.sel(rel_times=-23),
               levels=np.linspace(-3,3,13), cmap='RdBu', extend='both')
    ax.contour(tmp.rel_times, tmp.level, tmp.T-tmp.T.sel(rel_times=-23), 
               levels=np.linspace(-3,3,13), colors=['k'], linewidths=0.5)
    plt.colorbar(cf, ax=ax)
    if i == 0:
        ax.set_title('vertical profile of $\Delta$ $\Theta_{e}$ (before t=0 of MCSs)',fontsize=10)
    ax.set_xlabel('Time prior to init. (hr)')   
    ax.set_ylabel('Pressure (hPa)')
    ax.set_yticks([1000,850,500,200])
    ax.set_ylim([1000,200])
```

```python
# plot evolution of equivalent potential temperature, saturated

fig,(ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(6,6))

for ax,i in zip([ax1,ax2,ax3,ax4],range(4)):
    
    q = data_mcsenvs['Q'].sel(tracks=idx_quantile[i].values, rel_times=slice(-23,0)).mean('tracks')
    T = data_mcsenvs['T'].sel(tracks=idx_quantile[i].values, rel_times=slice(-23,0)).mean('tracks')
    qs = qs_calc(T)
    tmp = theta_e_calc(T, qs) # thetae profile
    cf = ax.contourf(tmp.rel_times, tmp.level, tmp.T-tmp.T.sel(rel_times=-23),
               levels=np.linspace(-2,2,11), cmap='RdBu', extend='both')
    ax.contour(tmp.rel_times, tmp.level, tmp.T-tmp.T.sel(rel_times=-23), 
               levels=np.linspace(-2,2,11), colors=['k'], linewidths=0.5)
    plt.colorbar(cf, ax=ax)
    if i == 0:
        ax.set_title('vertical profile of $\Delta$ $\Theta_{e, sat}$ (before t=0 of MCSs)',fontsize=10)
    ax.set_xlabel('Time prior to init. (hr)')   
    ax.set_ylabel('Pressure (hPa)')
    ax.set_yticks([1000,850,500,200])
    ax.set_ylim([1000,200])
```
