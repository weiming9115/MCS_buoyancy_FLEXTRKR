```python
def data_tracks_add_LSflag(data_tracks):
    """
    add landsea flag
    """

    lsflag_array = np.zeros((len(data_tracks.tracks),len(data_tracks.times)))*np.nan
    # ERA-5 lat-lon coordinate, 0.25-deg.
    data_lsmask = xr.open_dataset('/neelin2020/RGMA_feature_mask/ERA5_LandSeaMask_regrid.nc4')
    landseamask = data_lsmask.landseamask

    for (n,track) in enumerate(data_tracks.tracks.values):

        meanlon = data_tracks.sel(tracks=track).meanlon.values
        meanlat = data_tracks.sel(tracks=track).meanlat.values

        for t,(lon, lat) in enumerate(zip(meanlon, meanlat)):

            if np.isnan(lon) == 0: # not NaN

                if (lon < 0):
                    lon = lon + 360 # converting to 0-360
                
                landsea_flag = landseamask.sel(longitude=lon, latitude=lat, method='nearest').values

                if landsea_flag == 100: # ocean area (longitude, latitude)
                    lsflag_array[n,t] = 0
                else: # land area
                    lsflag_array[n,t] = 1
                    
    # write a new variable
    data_tracks['landsea_flag'] = data_tracks.meanlon*0 + lsflag_array
    data_tracks['landsea_flag'] = data_tracks['landsea_flag'].assign_attrs(units="0 = ocean; 1 = land", description="landsea flag")

    return data_tracks
```

```python
year = 2001 # get year integer

# data directoies
dir_mcs_track = Path('/scratch/wmtsai/temp_mcs/mcs_stats/mcs_tracks_non2mcs')
# read data
data_track = xr.open_dataset(dir_mcs_track / 'mcs_tracks_non2mcs_{}.tropics30NS.extend.nc'.format(year))
```

```python
data_addflag = data_tracks_add_LSflag(data_track)
```

#### check if landsea flag is correctly labeled

```python
# check total tracks over land and oceans during 2001-2020
year_list = np.arange(2001,2021)
track_num_land = 0
track_num_ocean = 0

dir_mcs_track = Path('/scratch/wmtsai/temp_mcs/mcs_stats/mcs_tracks_non2mcs')

for year in year_list:
    data_track = xr.open_dataset(dir_mcs_track / 'mcs_tracks_non2mcs_{}.tropics30NS.extend.nc'.format(year))
    idt_mcs_init = data_track.idt_mcs_init
    ls_flag = data_track.landsea_flag.sel(times=idt_mcs_init)
    track_num_land += len(np.where(ls_flag == 1)[0])
    track_num_ocean += len(np.where(ls_flag == 0)[0])
    
print('Tracks, Ocean: {}'.format(track_num_ocean))
print('Tracks, Land: {}'.format(track_num_land))
```

```python
year = 2002
dir_mcs_track = Path('/scratch/wmtsai/temp_mcs/mcs_stats/mcs_tracks_non2mcs')
data_track = xr.open_dataset(dir_mcs_track / 'mcs_tracks_non2mcs_{}.tropics30NS.extend.nc'.format(year))
```

```python
#ls_flag = data_track.landsea_flag.isel(times=0)
idt_mcs_init = data_track.idt_mcs_init
ls_flag = data_track.landsea_flag.sel(times=idt_mcs_init)
```

```python
idx_land = np.where(ls_flag == 1)[0] # land
idx_ocean = np.where(ls_flag == 0)[0] # ocean
mean_lat_land = data_track.meanlat.isel(tracks=idx_land, times=0)
mean_lon_land = data_track.meanlon.isel(tracks=idx_land, times=0)
mean_lat_ocean = data_track.meanlat.isel(tracks=idx_ocean, times=0)
mean_lon_ocean = data_track.meanlon.isel(tracks=idx_ocean, times=0)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,3),subplot_kw={'projection': ccrs.PlateCarree()})

ax.coastlines(color='k',linewidth=0.7)
ax.add_feature(cfeat.LAND,zorder=3,edgecolor='grey',facecolor='none',linewidth=0.3) # maskout LAND
ax.add_feature(cfeat.BORDERS,zorder=3,edgecolor='grey',facecolor='none',linewidth=0.3)
ax.add_feature(cfeat.STATES,zorder=3,edgecolor='grey',facecolor='none',linewidth=0.3) 

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='k', alpha=0.4, linestyle=':')
gl.xlabels_top = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabel_style = {'size': 9, 'color': 'k'}
gl.ylabel_style = {'size': 9, 'color': 'k'}

for lat, lon in zip(mean_lat_land, mean_lon_land):
    ax.plot(lon, lat, marker='x', ms=6)
    
ax.set_title('Examples of Tropical MCSs over Land, Init., {}'.format(year),fontsize=10)
plt.show()
```
