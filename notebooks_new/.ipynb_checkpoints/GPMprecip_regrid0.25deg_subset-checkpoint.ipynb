```python
import os
import sys
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
from datetime import datetime
from pathlib import Path

import cartopy.crs as ccrs
import cartopy.feature as cfeat
from cartopy.util import add_cyclic_point
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
year_list = np.arange(2002,2020)

for year in year_list:
    data_dir = Path('/scratch/wmtsai/GPM_feature_precip/{}'.format(year))
    files = sorted(list(data_dir.glob('GPM_feature_merged_*_v4.nc')))
    for file in files:
        month = str(file.name)[-8:-6]
        data = xr.open_dataset(file)
        data = data.sel(longitude=slice(0,359.75))
        
        # write out only GPM-precip
        precip = data.precipitationCal
        precip.to_netcdf(data_dir / 'GPM-IMERG_precip_{}_{}.0.25deg.nc'.format(year,month))
        print('GPM-IMERG_precip_{}_{}.0.25deg.nc ... processed'.format(year,month))
```
