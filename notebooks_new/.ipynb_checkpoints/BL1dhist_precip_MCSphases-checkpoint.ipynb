```python
# load unconditional, BL-precip data: BL1dhist_precip_writeout
out_dir = Path('/scratch/wmtsai/temp_mcs/output_stats/BLtot_precip_hist/')

data= xr.open_dataset(out_dir / 'hist1d_BL_climatology.ocean.nc')
samples = data.sum('phase').samples
prec_gpm_sum = data.sum('phase').prec_gpm_sum
pdf_uc_ocean = samples/samples.sum()
cond_prec_uc_ocean = prec_gpm_sum/samples
cond_prec_uc_ocean = cond_prec_uc_ocean.where(samples > 100)

data= xr.open_dataset(out_dir / 'hist1d_BL_climatology.land.nc')
samples = data.sum('phase').samples
prec_gpm_sum = data.sum('phase').prec_gpm_sum
pdf_uc_land = samples/samples.sum()
cond_prec_uc_land = prec_gpm_sum/samples
cond_prec_uc_land = cond_prec_uc_land.where(samples > 100)
```

```python
%%time
out_dir = Path('/scratch/wmtsai/temp_mcs/output_stats/BLtot_precip_hist/')

year_list = np.arange(2001,2021)

for n,year in enumerate(year_list):
    
    data = xr.open_dataset(out_dir / 'hist1d_BL_dtype.{}.ocean.3deg.TempCorr0.5.nc'.format(year))
    tmp = data.sum(('tracks','duration_type')).sel(bins_tot=slice(-29.75,10))
    
    if n == 0: 
        data_all_ocean = tmp
    else:
        data_all_ocean += tmp
```

```python
fig,ax = plt.subplots(1,1,figsize=(3.5,3.5))

ax2 = ax.twinx()
ccolors = ['k','orange','r','limegreen','b']

for phase,color in zip(data_all_ocean.phase.values[1:], ccolors):
    
    prec_gpm_sum = data_all_ocean.sel(area_type='mcs',phase=phase).prec_gpm_sum
    samples = data_all_ocean.sel(area_type='mcs',phase=phase).samples
    cond_prec_mcs = prec_gpm_sum/samples
    cond_prec_mcs = cond_prec_mcs.where(samples > 100)
    
    ax.plot(9.8/(340*3)*data_all_ocean.bins_tot, cond_prec_mcs, '*', ms=4, label=phase, color=color)
    ax2.plot(9.8/(340*3)*data_all_ocean.bins_tot, samples/samples.sum(), ':', lw=1.5, color=color)

# reference at zero buoyancy
ax.vlines(x=0,ymin=0,ymax=12,color='grey',lw=1,zorder=-1)

# reference for all grids, presenting climatgoloy 0.25deg BL-precip
#ax.plot(9.8/(340*3)*cond_prec_uc_ocean.bins_tot, cond_prec_uc_ocean, '-',color='grey', alpha=0.5, lw=2)
#ax2.plot(9.8/(340*3)*cond_prec_uc_ocean.bins_tot, pdf_uc_ocean, '-', color='grey', alpha=0.5, lw=2)

# subset for zoom-in around 0
ins = ax.inset_axes([0.12,0.22,0.4,0.3])
for phase,color in zip(data_all_ocean.phase.values[1:], ccolors):
    samples = data_all_ocean.sel(area_type='mcs',phase=phase).samples
    ins.plot(9.8/(340*3)*data_all_ocean.bins_tot, samples/samples.sum(), ':', lw=1.5, color=color)
ins.vlines(x=0,ymin=0,ymax=12,color='grey',lw=1,zorder=-1)
ins.set_xlim([-0.015,0.025])    
ins.set_ylim([0,0.025])
ins.tick_params(labelsize=6)
    
ax.legend(fontsize=9, loc=2, prop=arial_reg)
ax.grid(lw=0.5,ls=':')
ax.set_xlim([-0.3,0.05])
ax2.set_ylim([0,0.06])
ax.set_ylim([0,12])
ax.set_ylabel('Precipitation (mm/hr)',fontsize=10.5, font=arial_reg)
ax2.set_ylabel('PDF',fontsize=10, font=arial_reg)
ax.set_xlabel('B$_{L}$ (m/s$^{2}$)',fontsize=10.5, font=arial_reg)
ax.set_title('Conditional precipitation for MCSs, Ocean \n MCS grids (samples > 100, all tracks)',fontsize=9,
            font=arial_reg)

ax.tick_params(labelsize=9)
ax2.tick_params(labelsize=9)

#fig.savefig('/scratch/wmtsai/temp_mcs/fig/BL_precip_relationship_alltracks_ocean_MCS.pdf',
#          bbox_inches='tight', transparent=False)
```

```python
fig,ax = plt.subplots(1,1,figsize=(3.5,4))

ccolors = ['k','orange','r','limegreen','b']
samples_allphases = data_all_ocean.sel(area_type='mcs').isel(phase=slice(1,6)).samples.sum('phase')
prec_gpm_allphases = data_all_ocean.sel(area_type='mcs').isel(phase=slice(1,6)).prec_gpm_sum.sum('phase')
prec_gpm_cond = prec_gpm_allphases/samples_allphases

set_width=0.0012

for bt in data_all_ocean.bins_tot.values:

    prec_contrib = []
    for phase,ccolor in zip(data_all_ocean.phase.values[1:], ccolors):

        prec_gpm_sum = data_all_ocean.sel(bins_tot=bt,area_type='mcs',phase=phase).prec_gpm_sum.values
        prec_portion = (prec_gpm_sum/prec_gpm_allphases.sel(bins_tot=bt))*prec_gpm_cond.sel(bins_tot=bt)
        prec_contrib.extend([prec_portion.values])

    y1 = np.sort(prec_contrib)[0] # rank 1st 
    y2 = np.sort(prec_contrib)[1]
    y3 = np.sort(prec_contrib)[2]
    y4 = np.sort(prec_contrib)[3]
    y5 = np.sort(prec_contrib)[4]

    c1 = ccolors[np.argsort(prec_contrib)[0]]
    c2 = ccolors[np.argsort(prec_contrib)[1]]
    c3 = ccolors[np.argsort(prec_contrib)[2]]
    c4 = ccolors[np.argsort(prec_contrib)[3]]
    c5 = ccolors[np.argsort(prec_contrib)[4]]
    
    # plot bars in stack manner
    ax.bar(9.8/(340*3)*bt, y1, color=c1, width=set_width)
    ax.bar(9.8/(340*3)*bt, y2, bottom=y1, color=c2, width=set_width)
    ax.bar(9.8/(340*3)*bt, y3, bottom=y1+y2, color=c3, width=set_width)
    ax.bar(9.8/(340*3)*bt, y4, bottom=y1+y2+y3, color=c4, width=set_width)    
    ax.bar(9.8/(340*3)*bt, y5, bottom=y1+y2+y3+y4, color=c5, width=set_width)    

legend_elements = [Line2D([0], [0], color='k', lw=4, label='Initial'),
                   Line2D([0], [0], color='orange', lw=4, label='Grow'),
                   Line2D([0], [0], color='r', lw=4, label='Mature'),
                   Line2D([0], [0], color='limegreen', lw=4, label='Decay'),
                   Line2D([0], [0], color='b', lw=4, label='End')]
ax.legend(handles=legend_elements,fontsize=9, loc=2, prop=arial_reg)

# add precip conditioanl 
prec_gpm_allphases = data_all_ocean.sel(area_type='mcs').isel(phase=slice(1,6)).prec_gpm_sum.sum('phase')
prec_gpm_cond = prec_gpm_allphases/samples_allphases
ax.plot(9.8/(340*3)*data_all_ocean.bins_tot, prec_gpm_cond+0.1, '*',ms=4, color='k',
        alpha=0.5, zorder=3) 

# add samples info
samples_allphases = data_all_ocean.sel(area_type='mcs').isel(phase=slice(1,6)).samples.sum('phase')
ax2 = ax.twinx()
ax2.plot(9.8/(340*3)*data_all_ocean.bins_tot, samples_allphases, color='grey', lw=2, alpha=0.8, label=phase, zorder=2)
ax2.set_ylabel('Number of samples', color='gray', fontsize=10.5, font=arial_reg)

# add axis in unit of Kelvin
ax_up = ax.twiny()
ax_up.plot(9.8/(340*3)*data_all_ocean.bins_tot, samples_allphases, color='None', lw=2, alpha=0.8, label=phase, zorder=2)

bins_tot_vel = 9.8/(340*3)*data_all_ocean.bins_tot
bins_tot_degK = data_all_ocean.bins_tot
f = interpolate.interp1d(bins_tot_degK, bins_tot_vel)
xnew = np.arange(-25,5,5) # kelvin to be interpolated into
ynew = f(xnew) # location 
ax_up.set_xticks(ticks=ynew, labels=xnew)
ax_up.xaxis.set_tick_params(pad=0)
ax_up.tick_params(labelsize=9)

ax.set_ylim([0,9])
ax2.set_ylim([0,2.2e6])
ax.set_ylabel('Precip. contribution',fontsize=10.5, font=arial_reg)
ax.set_xlabel('B$_{L}$ (m/s$^{2}$)',fontsize=10.5, font=arial_reg)
ax.set_title('Contributions to precip, MCS grids, Ocean \n (affected by cloud size at the phase)',fontsize=9,
            font=arial_reg)
ax.set_ylabel('GPM precipitaiton (mm/hr)',fontsize=10.5)
ax.tick_params(labelsize=9)
ax.grid(lw=0.5,ls=':')

ax.set_xlim([-0.3,0.05])
ax_up.set_xlim([-0.3,0.05]);

# subset for zoom-in around negative buoyancy
ins = ax.inset_axes([0.12,0.22,0.4,0.3])
for bt in data_all_ocean.bins_tot.values:

    prec_contrib = []
    for phase,ccolor in zip(data_all_ocean.phase.values[1:], ccolors):

        prec_gpm_sum = data_all_ocean.sel(bins_tot=bt,area_type='mcs',phase=phase).prec_gpm_sum.values
        prec_portion = (prec_gpm_sum/prec_gpm_allphases.sel(bins_tot=bt))*prec_gpm_cond.sel(bins_tot=bt)
        prec_contrib.extend([prec_portion.values])

    y1 = np.sort(prec_contrib)[0] # rank 1st 
    y2 = np.sort(prec_contrib)[1]
    y3 = np.sort(prec_contrib)[2]
    y4 = np.sort(prec_contrib)[3]
    y5 = np.sort(prec_contrib)[4]

    c1 = ccolors[np.argsort(prec_contrib)[0]]
    c2 = ccolors[np.argsort(prec_contrib)[1]]
    c3 = ccolors[np.argsort(prec_contrib)[2]]
    c4 = ccolors[np.argsort(prec_contrib)[3]]
    c5 = ccolors[np.argsort(prec_contrib)[4]]
    
    # plot bars in stack manner
    ins.bar(9.8/(340*3)*bt, y1, color=c1, width=set_width)
    ins.bar(9.8/(340*3)*bt, y2, bottom=y1, color=c2, width=set_width)
    ins.bar(9.8/(340*3)*bt, y3, bottom=y1+y2, color=c3, width=set_width)
    ins.bar(9.8/(340*3)*bt, y4, bottom=y1+y2+y3, color=c4, width=set_width)    
    ins.bar(9.8/(340*3)*bt, y5, bottom=y1+y2+y3+y4, color=c5, width=set_width)    

ins.set_xlim([-0.2,-0.05])    
ins.set_ylim([0,1])
ins.tick_params(labelsize=6)

fig.savefig('/scratch/wmtsai/temp_mcs/fig/precip_cond_contribution_phases.ocean.ranked.TempCorr0.5.png',dpi=300,
         bbox_inches='tight',transparent=False)
```

```python
# load climatology 3-deg BL
data = xr.open_dataset('/scratch/wmtsai/temp_mcs/output_stats/era5_BL_3deg_2Dhistogram.wbwl.landsea.2002-2014.nc')
data = data.sum(('year','month'))

bins_cape = data.BL_cape_bins
bins_subsat = data.BL_subsat_bins
bins_tot = np.arange(-30,10,0.5)
samples_clim_oce = np.copy(bins_tot)
samples_clim_lnd = np.copy(bins_tot)

for cape in bins_cape.values:
    for subsat in bins_subsat.values:
        BL_tot = cape - subsat # (K) with wb, wl weighting
        idx = np.argmin(abs(BL_tot - bins_tot))
        samples_clim_oce[idx] += data.sel(BL_cape_bins=cape, BL_subsat_bins=subsat,surface_type='ocean').samples
        samples_clim_lnd[idx] += data.sel(BL_cape_bins=cape, BL_subsat_bins=subsat,surface_type='land').samples
```
