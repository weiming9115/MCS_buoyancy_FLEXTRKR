```python
import os
import sys
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
from datetime import datetime
from pathlib import Path
from matplotlib.lines import Line2D
from matplotlib.patches import Patch
from matplotlib.patches import Rectangle
from matplotlib.patches import FancyArrowPatch

import cartopy.crs as ccrs
import cartopy.feature as cfeat
from cartopy.util import add_cyclic_point
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
mcsstats_dir = Path('/scratch/wmtsai/temp_mcs/mcs_stats')
mcsmask_dir = Path('/neelin2020/mcs_flextrkr/20200101.0000_20210101.0000/')
mcsenv_dir = Path('/scratch/wmtsai/temp_mcs/mcs_stats/envs_track/2020/tropics_extend')
```

```python
data_mcsstats = xr.open_dataset(mcsstats_dir / 'mcs_tracks_final_extc_20200101.0000_20210101.0000.nc')
meanlat = data_mcsstats.meanlat.isel(times=0)
idx_track = np.where(abs(meanlat) < 30)[0]
data_mcsstats = data_mcsstats.isel(tracks=idx_track)
```

```python
fig, ax = plt.subplots(1,1,figsize=(6,4))

track_number = 200

ax2 = ax.twinx()
ax3 = ax.twinx()
data_mcsstats.isel(tracks=track_number).ccs_area.plot(ax=ax, marker='o', color='k')
data_mcsstats.isel(tracks=track_number).mcs_status.plot(ax=ax2, marker='o', color='b')
data_mcsstats.isel(tracks=track_number).total_rain.plot(ax=ax3, marker='o', color='r')

ax.set_ylabel('CCS Area (km$^2$)',fontsize=10)
ax2.set_ylabel('MCS status (0 or 1)',fontsize=10)
ax3.set_ylabel('Total rain (mm/hr)',fontsize=10)
ax.set_xlabel('hour after the detection of CCS')

ax2.spines.right.set_position(("axes", 1))
ax3.spines.right.set_position(("axes", 1.15))
ax2.tick_params(axis='y', colors='blue')
ax3.tick_params(axis='y', colors='red')

ax.grid(lw=0.5, ls=':')
plt.show()
```

```python
data_mcsmask = xr.open_dataset(mcsmask_dir / 'mcstrack_20201001_0630.nc').isel(time=0)
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,3),subplot_kw={'projection': ccrs.PlateCarree(160)})

ax.coastlines(color='k',linewidth=0.5)
ax.add_feature(cfeat.LAND,zorder=3,edgecolor='grey',facecolor='none',linewidth=0.3) # maskout LAND
ax.add_feature(cfeat.BORDERS,zorder=3,edgecolor='grey',facecolor='none',linewidth=0.3)
ax.spines[:].set_visible(False)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                 linewidth=0.1, color='grey', alpha=0.4, linestyle=':')
gl.xlabels_top = False
gl.xlabels_bottom = False
gl.ylabels_right = False
gl.xformatter = LONGITUDE_FORMATTER
gl.yformatter = LATITUDE_FORMATTER
gl.xlabel_style = {'size': 7, 'color': 'grey'}
gl.ylabel_style = {'size': 7, 'color': 'grey'}

cp = ax.pcolormesh(data_mcsmask.lon, data_mcsmask.lat, data_mcsmask.cloudtracknumber_nomergesplit, 
             transform=ccrs.PlateCarree(), cmap='tab20')
cbar = plt.colorbar(cp, shrink=0.6)
cbar.set_label('Track number',fontsize=10)
plt.show()
```
