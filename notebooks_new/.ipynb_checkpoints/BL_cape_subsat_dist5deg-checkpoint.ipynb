```python
feat_dir = Path('/scratch/wmtsai/temp_mcs/mcs_stats/mcs_tracks_non2mcs/tracks_area_mean')
```

```python
def bins_duration_samples(BL_var, mcs_phase, duration_bins, BL_bins, factor):
    
    dx = BL_bins[1]-BL_bins[0]
    dy = duration_bins[1]-duration_bins[0]    
    bins_samples = np.zeros((len(BL_bins)-1, len(duration_bins)-1))

    year_list = np.arange(2001,2020)
    for year in year_list:
        data = xr.open_dataset(feat_dir / 'featstats_tracks_non2mcs_{}.tropics30NS.nc'.format(year))
        BL_1d = factor*data[BL_var].sel(mcs_phase=mcs_phase).values
        duration_1d = data.mcs_duration

        for i in range(len(BL_bins)-1):
            idx = np.where(np.logical_and(BL_1d >= BL_bins[i], BL_1d < BL_bins[i+1]))[0]
            for j in range(len(duration_bins)-1):
                idy = np.where(np.logical_and(duration_1d >= duration_bins[j], duration_1d < duration_bins[j+1]))[0]
                idx_com = np.intersect1d(idx,idy)
                bins_samples[i,j] += len(idx_com)
                
    # writeout as xarray
    bins_samples_xr = xr.Dataset(data_vars=dict(samples=(['BL_bins','duration_bins'], bins_samples)),
                                 coords=dict(duration_bins=(['duration_bins'],duration_bins[:-1]+dy/2),
                                             BL_bins=(['BL_bins'], BL_bins[:-1]+dx/2)))
    
    return bins_samples_xr
```

```python
def plot_BL_duration(bins_samples_xr, xlim, x_label, title):
    
    # get marginal distribution of duration 
    duration_bins = bins_samples_xr.duration_bins
    BL_bins = bins_samples_xr.BL_bins

    margin_prob_duration = np.zeros((len(BL_bins),len(duration_bins)))
    for i in range(len(BL_bins)):
        margin_prob_duration[i,:] = bins_samples_xr.samples.values[i,:]/np.sum(bins_samples_xr.samples.values[i,:])  

    # conditional mean of MCS duration at a given BL,tot
    cond_duration = np.dot(duration_bins, margin_prob_duration.T)

    # 95 percentile
    cumul_prob = np.cumsum(margin_prob_duration, axis=1)
    duration_95th = np.zeros(cumul_prob.shape[0])*np.nan
    for i in range(cumul_prob.shape[0]):
        idx = np.argmin(abs(cumul_prob[i,:] - 0.95))
        if idx > 0:
            duration_95th[i] = duration_bins[idx]
    
    # plotting section
    fig, ax = plt.subplots(1,1,figsize=(5,3.5))
    cp = ax.pcolormesh(BL_bins, duration_bins, margin_prob_duration.T, vmin=0.001, vmax=0.2,
                  cmap='jet')
    cp.cmap.set_under('w')
    cbar = plt.colorbar(cp,ax=ax,shrink=0.7,extend='min',pad=0.18)

    # plot samples 
    ax2 = ax.twinx()
    ax2.plot(BL_bins, bins_samples_xr.samples.sum('duration_bins'), 'grey', lw=1.5)

    # plot conditional mean of MCS duration 
    ax.plot(BL_bins, cond_duration, 'm', lw=1.5)
    # 95 percentile
    ax.plot(BL_bins, duration_95th, 'r', lw=1.5)

    ax.set_xlim([xlim[0],xlim[1]])
    ax.set_xlabel(x_label,fontsize=10)
    ax.set_ylabel('MCS duration (hour)',fontsize=10)
    ax.set_title(title,fontsize=10.5);

    ax2.set_ylim([0,9000])
    ax2.set_ylabel('samples',fontsize=10)
```

#### growing phase

```python
%%time
# BL_TOT_amean, 5-deg at Grow
bins_samples_tot_xr = bins_duration_samples(BL_var='BL_TOT_amean',mcs_phase='Init'
                             ,duration_bins=np.arange(5,80,5), BL_bins=np.arange(-16,10,0.5), factor=(340*3)/9.8)
# BL_CAPE_amean, 5-deg at Grow
bins_samples_cape_xr = bins_duration_samples(BL_var='BL_CAPE_amean',mcs_phase='Init'
                             ,duration_bins=np.arange(5,80,5), BL_bins=np.arange(-16,10,0.5), factor=(340*3)/9.8/0.4)
# BL_SUBSAT_amean, 5-deg at Grow
bins_samples_subsat_xr = bins_duration_samples(BL_var='BL_SUBSAT_amean',mcs_phase='Init'
                             ,duration_bins=np.arange(5,80,5), BL_bins=np.arange(-16,10,0.5), factor=-(340*3)/9.8/0.6)
```

```python
plot_BL_duration(bins_samples_tot_xr, xlim=[-15,10], x_label='B$_{L,tot}$ (K)',
                title='5-deg, BL_TOT, Grow (2001-2019)')
```
