```python
era5_env_dir = Path('/neelin2020/mcs_flextrkr/era5_envs/')
featstats_dir = Path('/scratch/wmtsai/temp_mcs/mcs_stats/mcs_tracks_non2mcs/tracks_area_mean')
mcsstats_dir = Path('/scratch/wmtsai/temp_mcs/mcs_stats/mcs_tracks_non2mcs')
```

```python
%%time
year_list = np.arange(2001,2021)

# bins for BL_CAPE and BL_SUBSAT
bins_cape = np.linspace(-20,30,101)
bins_subsat = np.linspace(-20,30,101)
dbl_cape = (bins_cape[1]-bins_cape[0])/2
dbl_subsat= (bins_subsat[1]-bins_subsat[0])/2

samples = np.zeros((3, 2, len(bins_cape)-1, len(bins_subsat)-1)) # (CCS/Init/Mature, surface_type, shear_type, BL-cape, BL-subsat) 
shear_amp_sum = np.zeros((3, 2, 4, len(bins_cape)-1, len(bins_subsat)-1))
shear_amp_num = np.copy(shear_amp_sum)

# BL parameters from m/s^2 to K    
delta_pl=400
delta_pb=100
wb=(delta_pb/delta_pl)*np.log((delta_pl+delta_pb)/delta_pb)
wl=1-wb

for year in year_list:
    
    print('processing year: {}'.format(year))

    data_era5_env = xr.open_dataset(era5_env_dir / 'mcs_era5_mean_envs_{}0101.0000_{}0101.0000.nc'.format(year,year+1))
    data_featstats = xr.open_dataset(featstats_dir / 'featstats_tracks_non2mcs_{}.tropics30NS.extend.nc'.format(year))
    data_mcsstats = xr.open_dataset(mcsstats_dir / 'mcs_tracks_non2mcs_{}.tropics30NS.extend.nc'.format(year))
    # extract the corresponding non2MCS tracks from the original tracks
    tracks_sel = data_featstats.tracks
    data_era5_env = data_era5_env.sel(tracks=tracks_sel)
    
    for p,(phase,var_string) in enumerate(zip(['CCS','Init','Mature']
                                              ,['idt_ccs_init','idt_mcs_init','idt_mcs_mature'])):
        
        BL_CAPE_amean = (340*3)/9.81/wb*data_featstats.BL_CAPE_3degmean.sel(mcs_phase=phase).values
        BL_SUBSAT_amean = (340*3)/9.81/wl*data_featstats.BL_SUBSAT_3degmean.sel(mcs_phase=phase).values
        
        for i in range(len(bins_cape)-1):
            idx = np.where(np.logical_and(BL_CAPE_amean >= bins_cape[i], BL_CAPE_amean < bins_cape[i+1]))[0]
            for j in range(len(bins_subsat)-1):
                idy = np.where(np.logical_and(BL_SUBSAT_amean >= bins_subsat[j], BL_SUBSAT_amean < bins_subsat[j+1]))[0]
                idx_com = np.intersect1d(idx,idy)
                
                if len(idx_com) > 0: # if getting samples                    
                    # get shear info at the corresponding times
                    idt_phase = data_mcsstats[var_string].isel(tracks=idx_com) 

                    for idc,idt in zip(idx_com, idt_phase):                    
                        if ~np.isnan(data_era5_env.shear_mag_0to2km.isel(tracks=idc).sel(rel_times=idt)) == 1:
                            ls_flag = int(data_mcsstats.landsea_flag.isel(tracks=idc).sel(times=idt))
                            shear_amp_sum[p,ls_flag,0,i,j] += data_era5_env.shear_mag_0to2km.isel(tracks=idc).sel(rel_times=idt)
                            shear_amp_num[p,ls_flag,0,i,j] += 1
                        if ~np.isnan(data_era5_env.shear_mag_0to4km.isel(tracks=idc).sel(rel_times=idt)) == 1:
                            ls = int(data_mcsstats.landsea_flag.isel(tracks=idc).sel(times=idt))
                            shear_amp_sum[p,ls_flag,1,i,j] += data_era5_env.shear_mag_0to4km.isel(tracks=idc).sel(rel_times=idt)
                            shear_amp_num[p,ls_flag,1,i,j] += 1                        
                        if ~np.isnan(data_era5_env.shear_mag_4to8km.isel(tracks=idc).sel(rel_times=idt)) == 1:
                            ls = int(data_mcsstats.landsea_flag.isel(tracks=idc).sel(times=idt))
                            shear_amp_sum[p,ls_flag,2,i,j] += data_era5_env.shear_mag_4to8km.isel(tracks=idc).sel(rel_times=idt)
                            shear_amp_num[p,ls_flag,2,i,j] += 1                       
                        if ~np.isnan(data_era5_env.shear_mag_6to10km.isel(tracks=idc).sel(rel_times=idt)) == 1:
                            ls = int(data_mcsstats.landsea_flag.isel(tracks=idc).sel(times=idt))
                            shear_amp_sum[p,ls_flag,3,i,j] += data_era5_env.shear_mag_6to10km.isel(tracks=idc).sel(rel_times=idt)
                            shear_amp_num[p,ls_flag,3,i,j] += 1
                        
# write into xarray dataset
ds = xr.Dataset(data_vars=dict(shear_amp_sum=(['mcs_phase','surface_type','shear_type','bins_cape','bins_subsat'],shear_amp_sum),
                               shear_amp_num=(['mcs_phase','surface_type','shear_type','bins_cape','bins_subsat'],shear_amp_num)),
                coords=dict(mcs_phase=(['mcs_phase'],['CCS','Init','Mature']),
                            surface_type=(['surface_type'],['ocean','land']),
                            shear_type=(['shear_type'],['0to2km','0to4km','4to8km','6to10km']),
                            bins_cape=(['bins_cape'],bins_cape[:-1]),
                            bins_subsat=(['bins_subsat'],bins_subsat[:-1])))
```

```python
fig,(ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(11,2.5))

mcs_phase = 'CCS'
surface_type= 'ocean'

# at the CCS phase
for ax,shear_type in zip([ax1,ax2,ax3,ax4],['0to2km','0to4km','4to8km','6to10km']):
    #samples = ds.sel(mcs_phase='Init').samples
    shear_amp_sum = ds.sel(mcs_phase=mcs_phase,surface_type=surface_type,shear_type=shear_type).shear_amp_sum
    shear_amp_num = ds.sel(mcs_phase=mcs_phase,surface_type=surface_type,shear_type=shear_type).shear_amp_num
    
    shear_mean = shear_amp_sum/shear_amp_num
    dbl = 0.5
    shear_mean_interp = shear_mean.interp(bins_cape=np.arange(bins_cape[0],bins_cape[-1]+dbl,dbl),
                          bins_subsat=np.arange(bins_subsat[0],bins_subsat[-1]+dbl,dbl))
    shear_amp_num_interp = shear_amp_num.interp(bins_cape=np.arange(bins_cape[0],bins_cape[-1]+dbl,dbl),
                          bins_subsat=np.arange(bins_subsat[0],bins_subsat[-1]+dbl,dbl))
    
    shear_mean_interp = shear_mean_interp.where(shear_amp_num_interp > 20) # filtering out insufficient samples
    
    cp = ax.pcolormesh(shear_mean_interp.bins_subsat, shear_mean_interp.bins_cape,shear_mean_interp, cmap='jet', vmin=0, vmax=4, zorder=2)
    #ax.contour(shear_mean_interp.bins_subsat, shear_mean_interp.bins_cape, shear_mean_interp, levels=[2,2.5,3,3.5,4],colors=['m'],
    #          linewidths=0.5,zorder=3)
    #ax.contour(shear_amp_num.bins_subsat, shear_amp_num.bins_cape, shear_amp_num/shear_amp_num.sum(),levels=5,colors=['k'],
    #          linewidths=0.5,zorder=3)
    cbar = plt.colorbar(cp, ax=ax, shrink=0.7)
    cbar.set_label('amplitude (m/s)',fontsize=10)

    # referecne line of BL = 0K
    ax.plot(np.linspace(-2,10,13),np.linspace(-2,10,13), color='k', lw=1)
    
    ax.set_xlabel('$\hat{B}_{L,subsat}}$ (K)',fontsize=10)
    ax.set_ylabel('$\hat{B}_{L,cape}$ (K)',fontsize=10)
    ax.set_title(shear_type, fontsize=11)
    ax.set_ylim([-2,10])
    ax.set_xlim([-2,12])
    ax.set_yticks([0,2,4,6,8,10])
    ax.set_xticks([0,2,4,6,8,10])
    ax.grid(linestyle=':',lw=1,color='grey',zorder=1)
    
plt.suptitle('3-deg averages of BL & Shear, MCS phase: {}'.format(mcs_phase),fontsize=10, fontweight='bold',y=0.9)
plt.tight_layout()
```
