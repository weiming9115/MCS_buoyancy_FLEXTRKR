```python
%%time
year_list = np.arange(2001,2021)

# bins for BL_CAPE and BL_SUBSAT
bins_cape = np.linspace(-10,10,101)
bins_subsat = np.linspace(-5,15,101)
dbl_cape = (bins_cape[1]-bins_cape[0])/2
dbl_subsat= (bins_subsat[1]-bins_subsat[0])/2

samples = np.zeros((4, 2, len(bins_cape)-1, len(bins_subsat)-1)) 
# (CCS/Init/Mature, surface_type, shear_intensity, BL-cape, BL-subsat) 
var_sum = np.zeros((4, 2, 5, len(bins_cape)-1, len(bins_subsat)-1)) # variable = mcs duration here
var_num = np.copy(var_sum)

for year in year_list:
    
    print('processing year: {}'.format(year))

    data_era5_env = xr.open_dataset(era5_env_dir / 'mcs_era5_mean_envs_{}0101.0000_{}0101.0000.nc'.format(year,year+1))
    data_featstats = xr.open_dataset(featstats_dir / 'featstats_tracks_non2mcs_{}.tropics30NS.extend.nc'.format(year))
    data_mcsstats = xr.open_dataset(mcsstats_dir / 'mcs_tracks_non2mcs_{}.tropics30NS.extend.nc'.format(year))
    # extract the corresponding non2MCS tracks from the original tracks
    tracks_sel = data_featstats.tracks
    data_era5_env_sub = data_era5_env.sel(tracks=tracks_sel)
    
    for p,(phase,var_string) in enumerate(zip(['CCS','Init','Grow','Mature']
                                              ,['idt_ccs_init','idt_mcs_init','idt_mcs_grow','idt_mcs_mature'])):
        
        BL_CAPE_amean = data_featstats.BL_CAPE_mcs.sel(mcs_phase=phase).values # wb*BL_CAPE (K)
        BL_SUBSAT_amean = data_featstats.BL_SUBSAT_mcs.sel(mcs_phase=phase).values #wl*BL_SUBSAT (K)
        
        for i in range(len(bins_cape)-1):
            idx = np.where(np.logical_and(BL_CAPE_amean >= bins_cape[i], BL_CAPE_amean < bins_cape[i+1]))[0]
            for j in range(len(bins_subsat)-1):
                idy = np.where(np.logical_and(BL_SUBSAT_amean >= bins_subsat[j], BL_SUBSAT_amean < bins_subsat[j+1]))[0]
                idx_com = np.intersect1d(idx,idy)
                
                # go to data_era5_env and extract variables from the corresponding tracks and times
                
                if len(idx_com) > 0: # if getting samples                    
                    # get shear info at the corresponding times
                    idt_phase = data_mcsstats[var_string].isel(tracks=idx_com) # timestamp of each track

                    for idc,idt in zip(idx_com, idt_phase):                    
                        if ~np.isnan(data_era5_env_sub.shear_mag_0to2km.isel(tracks=idc).sel(rel_times=idt)) == 1:
                            
                            # land-sea flag
                            ls_flag = int(data_mcsstats.landsea_flag.isel(tracks=idc).sel(times=0))
                            
                            # shear intensity flag
                            shear_mag = data_era5_env_sub.shear_mag_0to2km.isel(tracks=idc).sel(rel_times=idt).values
                            
                            if shear_mag >= 0 and shear_mag < 2.5:
                                shear_flag = 0
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                                
                            if shear_mag >= 2.5 and shear_mag < 5:
                                shear_flag = 1
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                                
                            if shear_mag >= 5 and shear_mag < 7.5:
                                shear_flag = 2
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                                
                            if shear_mag >= 7.5 and shear_mag < 10:
                                shear_flag = 3
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                                
                            if shear_mag >= 10:
                                shear_flag = 4
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                        
# write into xarray dataset
ds_shear0to2km = xr.Dataset(data_vars=dict(var_sum=(['mcs_phase','surface_type','shear_type','bins_cape','bins_subsat'],var_sum),
                                           var_num=(['mcs_phase','surface_type','shear_type','bins_cape','bins_subsat'],var_num)),
                coords=dict(mcs_phase=(['mcs_phase'],['CCS','Init','Grow','Mature']),
                            surface_type=(['surface_type'],['ocean','land']),
                            shear_type=(['shear_type'],['0to2.5','2.5to5','5to7.5','7.5to10','10+']),
                            bins_cape=(['bins_cape'],bins_cape[:-1]),
                            bins_subsat=(['bins_subsat'],bins_subsat[:-1])))
```

```python
%%time
year_list = np.arange(2001,2021)

# bins for BL_CAPE and BL_SUBSAT
bins_cape = np.linspace(-20,30,101)
bins_subsat = np.linspace(-20,30,101)
dbl_cape = (bins_cape[1]-bins_cape[0])/2
dbl_subsat= (bins_subsat[1]-bins_subsat[0])/2

samples = np.zeros((3, 2, len(bins_cape)-1, len(bins_subsat)-1)) # (CCS/Init/Mature, surface_type, shear_intensity, BL-cape, BL-subsat) 
var_sum = np.zeros((3, 2, 5, len(bins_cape)-1, len(bins_subsat)-1)) # variable = mcs duration here
var_num = np.copy(var_sum)

# BL parameters from m/s^2 to K    
delta_pl=400
delta_pb=100
wb=(delta_pb/delta_pl)*np.log((delta_pl+delta_pb)/delta_pb)
wl=1-wb

for year in year_list:
    
    print('processing year: {}'.format(year))

    data_era5_env = xr.open_dataset(era5_env_dir / 'mcs_era5_mean_envs_{}0101.0000_{}0101.0000.nc'.format(year,year+1))
    data_featstats = xr.open_dataset(featstats_dir / 'featstats_tracks_non2mcs_{}.tropics30NS.extend.nc'.format(year))
    data_mcsstats = xr.open_dataset(mcsstats_dir / 'mcs_tracks_non2mcs_{}.tropics30NS.extend.nc'.format(year))
    # extract the corresponding non2MCS tracks from the original tracks
    tracks_sel = data_featstats.tracks
    data_era5_env = data_era5_env.sel(tracks=tracks_sel)
    
    for p,(phase,var_string) in enumerate(zip(['CCS','Init','Mature']
                                              ,['idt_ccs_init','idt_mcs_init','idt_mcs_mature'])):
        
        BL_CAPE_amean = (340*3)/9.81/wb*data_featstats.BL_CAPE_3degmean.sel(mcs_phase=phase).values
        BL_SUBSAT_amean = (340*3)/9.81/wl*data_featstats.BL_SUBSAT_3degmean.sel(mcs_phase=phase).values        
        
        for i in range(len(bins_cape)-1):
            idx = np.where(np.logical_and(BL_CAPE_amean >= bins_cape[i], BL_CAPE_amean < bins_cape[i+1]))[0]
            for j in range(len(bins_subsat)-1):
                idy = np.where(np.logical_and(BL_SUBSAT_amean >= bins_subsat[j], BL_SUBSAT_amean < bins_subsat[j+1]))[0]
                idx_com = np.intersect1d(idx,idy)
                
                # go to data_era5_env and extract variables from the corresponding tracks and times
                
                if len(idx_com) > 0: # if getting samples                    
                    # get shear info at the corresponding times
                    idt_phase = data_mcsstats[var_string].isel(tracks=idx_com) # timestamp of each track

                    for idc,idt in zip(idx_com, idt_phase):                    
                        if ~np.isnan(data_era5_env.shear_mag_4to8km.isel(tracks=idc).sel(rel_times=idt)) == 1:
                            
                            # land-sea flag
                            ls_flag = int(data_mcsstats.landsea_flag.isel(tracks=idc).sel(times=0))
                            
                            # shear intensity flag
                            shear_mag = data_era5_env.shear_mag_4to8km.isel(tracks=idc).sel(rel_times=idt).values
                            
                            if shear_mag >= 0 and shear_mag < 5:
                                shear_flag = 0
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                                
                            if shear_mag >= 5 and shear_mag < 10:
                                shear_flag = 1
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                                
                            if shear_mag >= 10 and shear_mag < 15:
                                shear_flag = 2
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                                
                            if shear_mag >= 15 and shear_mag < 20:
                                shear_flag = 3
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                                
                            if shear_mag >= 20:
                                shear_flag = 4
                                var_sum[p,ls_flag,shear_flag,i,j] += int(data_mcsstats.mcs_duration.isel(tracks=idc).values)
                                var_num[p,ls_flag,shear_flag,i,j] += 1
                        
# write into xarray dataset
ds_shear4to8km = xr.Dataset(data_vars=dict(var_sum=(['mcs_phase','surface_type','shear_type','bins_cape','bins_subsat'],var_sum),
                                           var_num=(['mcs_phase','surface_type','shear_type','bins_cape','bins_subsat'],var_num)),
                coords=dict(mcs_phase=(['mcs_phase'],['CCS','Init','Mature']),
                            surface_type=(['surface_type'],['ocean','land']),
                            shear_type=(['shear_type'],['0to5','5to10','10to15','15to20','20+']),
                            bins_cape=(['bins_cape'],bins_cape[:-1]),
                            bins_subsat=(['bins_subsat'],bins_subsat[:-1])))
```
